## Instagram influencer engagement rate by follower count, vs other platforms

```python
import pandas as pd                #import pandas
import matplotlib.pyplot as plt    #import matplotlib
import numpy as np

stats = pd.read_csv("D:\DataScience\Instagram DataAnalysis\ Marketing Statistics\MS11.csv", header=0, sep=",")
print(stats)

df = pd.DataFrame(stats)           #make dataframe
x = list(df.iloc[:, 0])            #get 0th col
y = list(df.iloc[:, 1])            #get 1st col
z = list(df.iloc[:, 2])
w = list(df.iloc[:, 3])

N = len(x)
ind = np.arange(N)
plt.figure(figsize=(15,7))
width = 0.3 

# Plotting
plt.bar(ind, y , width, label='Instagram')
plt.bar(ind + width, z, width, label='Tiktok')
plt.bar(ind + 2*width, w, width, label='Twitter')

plt.xticks(ind + width / 2, x)

plt.xlabel("Follower_count")
plt.ylabel("% of Follower_count")
plt.title("Instagram influencer engagement rate by follower count, vs other platforms")

# Finding the best position for legends and putting it
plt.legend(loc='best')
plt.show()
```
