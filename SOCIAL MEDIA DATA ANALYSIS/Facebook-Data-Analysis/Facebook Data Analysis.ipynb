## Total Facebook users


```python
import pandas as pd  # import pandas
import matplotlib.pyplot as plt     #import matplotlib

# reading the csv file 
stats = pd.read_csv("D:\DataScience\Facebook Data Analysis\FB1.csv", header=0, sep=",")
print("Table :\n")
print(stats)

#making a dataframe
df = pd.DataFrame(stats)

#get x list
x = list(df.iloc[:, 0])
#get y list
y = list(df.iloc[:, 1])
#colors of bars
c = ['r','r','r','r','g','g','g','g','b','b','b','b','k','k','k','k']

#plotting the bars
plt.figure(figsize=(30,20))
plt.title("Total Facebook users 2021*")          #select a title for bar
plt.xlabel("Monthly_active_users(millions)")     #select a x label
plt.ylabel("Quarter")                            #select a y label

plt.grid(axis='x', color='r')
plt.barh(x, y, color=c)                                  #plotting a horizantal graph
plt.show()                                      # show the bar
```

## Facebook users by region, millions


```python
import pandas as pd  # import pandas
import matplotlib.pyplot as plt     #import matplotlib
import numpy as np

# reading the csv file 
stats = pd.read_csv("D:\DataScience\Facebook Data Analysis\FB5.csv", header=0, sep=",")
print("Table :\n")
print(stats)

#making a dataframe
df = pd.DataFrame(stats)           #make dataframe
x = list(df.iloc[:, 0])            #get 0th col
y = list(df.iloc[:, 1])            #get 1st col
z = list(df.iloc[:, 2])
w = list(df.iloc[:, 3])
k = list(df.iloc[:, 4])

N = len(x)
ind = np.arange(N)
plt.figure(figsize=(21,10))
width = 0.2 

# Plotting
plt.grid(axis='y', color='g')
plt.bar(ind, y , width, label='US&Canada', color='r')
plt.bar(ind + width, z, width, label='Europe', color='b')
plt.bar(ind + 2*width, w, width, label='Asia_Pacific', color='k')
plt.bar(ind + 4*width, w, width, label='Rest_of_World', color='y')

plt.xticks(ind + width / 2, x)

plt.xlabel("Quarter")
plt.ylabel("Millions")
plt.title("Facebook users by region, millions")

# Finding the best position for legends and putting it
plt.legend(loc='best')
plt.show()
```
